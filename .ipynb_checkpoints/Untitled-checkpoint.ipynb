```python
#correlation heatmap to check for multi-correnality
numeric_data = df.select_dtypes(include=['number'])
plt.figure(figsize=(12,8))
sns.heatmap(numeric_data.corr(), annot=True,cmap='Greens')
plt.title("correlation heatmap")
plt.show()
```

```python
#feature engineering
cat_features= df.select_dtypes(include=['object']).columns
num_features= df.select_dtypes(exclude=['object']).columns
```

```python
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
numeric_transformer = StandardScaler()

preprocessor = ColumnTransformer(
    [
        ("StandardScaler", numeric_transformer, num_features),
        ("OneHotEncoder", categorical_transformer, cat_features)
    ],
    remainder='passthrough'
)
x_processed=preprocessor.fit_transform(df)
```

```python
print(df.columns.tolist())
```

```python
 df1=df[["lead_time", "booking_changes","hotel","adults","meal","country","previous_cancellations","deposit_type","customer_type","total_of_special_requests","reservation_status","adr"]]
```

```python
df1.plot(kind="scatter",
        x="lead_time",
        y="booking_changes",
        figsize=(10,7))
plt.show()
```

```python
#finding the optimal number of clusters with the elbow method
sum_of_sqr_dist = {}

for k in range (1,30):
    print(f"Running K-Means for k = {k}...")
    km = KMeans(n_clusters=k, init='k-means++', max_iter=1000, random_state=42, n_init=10)
    km.fit(x_processed)
    sum_of_sqr_dist[k] = km.inertia_
    print(f"Finished k = {k}. Inertia: {km.inertia_}")
```

```python
sns.pointplot(x=list(sum_of_sqr_dist.keys()), y=list(sum_of_sqr_dist.values()))
plt.xlabel("number of clusters k")
plt.ylabel("sum of square distances")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
model = KMeans(n_clusters=8,
               init='k-means++',
               max_iter=1000)
model.fit(x_processed)
```

```python
print("labels", model.labels_)
```

```python
print("centroids", model.cluster_centers_)
```

```python
centroids=model.cluster_centers_
```

```python
#calculating the silhouette score
cluster_labels = model.labels_

score = silhouette_score(x_processed, cluster_labels)

print(f"The Silhouette Score for k=8 is: {score}")
```

```python
df1_cluster=df1.copy()
df1_cluster["cluster"] = model.fit_predict(x_processed)
```
